# Import necessary libraries

```python
import pandas as pd
```

# Load the dataset

```python
df = pd.read_csv('KaggleV2-May-2016.csv')
df.head()
df.info()
df.isnull().sum()
```

# Check for duplicates

```python
df.duplicated().sum()
```

# Summary statistics for numerical columns


```python
df.describe()
```

# Remove invalid age entries (negative values)

```python
df = df[df['Age'] >= 0]
df
```

# Standardize column names (lowercase, underscores)

```python
df.columns = df.columns.str.strip().str.lower().str.replace("-", "_")
df
```

# Rename columns correctly

```python
df.rename(columns={"hipertension": "hypertension", "handcap": "handicap"}, inplace=True)
df
```

# Convert 'handicap' to binary (0 = no handicap, 1 = any handicap)

```python
df['handicap'] = df['handicap'].apply(lambda x: 1 if x > 0 else 0)
df
```

# Convert date columns to datetime objects

```python
df['scheduledday'] = pd.to_datetime(df['scheduledday'])
df['appointmentday'] = pd.to_datetime(df['appointmentday'])
df
```

# Encode 'No-show' column: Yes = 1 (missed), No = 0 (attended)

```python
df['no_show'] = df['no_show'].map({'Yes': 1, 'No': 0})
df
```

```python
df.drop(['patientid'], axis=1, inplace=True)
```

# Standardize Gender

```python
df['gender'] = df['gender'].str.strip().str.upper()
df['gender'] = df['gender'].replace({'M': 'Male', 'F': 'Female'})
```

# Capitalize neighborhood names properly

```python
df['neighbourhood'] = df['neighbourhood'].str.title()
```

```python
df.info()
df.describe()
df.head()
```

```python
df.to_csv("cleaned_medical_appointments.csv", index=False)
```
